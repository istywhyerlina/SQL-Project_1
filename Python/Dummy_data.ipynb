```python
from faker import Faker
import random
import re

n_user=100
fake = Faker('id_ID')

def generate_dummy_user(n):
    """
    A function to generate user data by Faker.

    Args
    n (int) : amount of users to be generated

    Return
    user_data (list)
    """
    user_data = []

    for i in range(1,n+1):
        pn=fake.phone_number()
        new_pn=re.sub(r'[^\w]', '', pn)
        created_date = fake.date_time_between_dates(datetime_start =datetime(2021, 1, 1, 0, 0, 0), datetime_end =datetime(2023, 12, 31, 23, 59, 59)).strftime("%Y-%m-%d %H:%M:%S")
        user_data.append({
            "user_id": i,
            "name": fake.name(),
            "email":fake.email(),
            "phone_number": new_pn,
            "city_id": random.choice(cities['city_id']),
            "date_created": created_date
        })

    return user_data




#define column for user table
user_cols = ['user_id', 'name','email' ,'phone_number', 'city_id','date_created']
#generate rows for user table
user_rows = generate_dummy_user(n_user)
user_df=pd.DataFrame(user_rows, columns=user_cols)
user_df
```

```python
n_ads=200
def generate_dummy_ads(n,user_rows):
    """
    A function for generating advertisment data.

    Args
    n (int) : desired number of ads to be generated
    users_data (list) : list of user id
    cars_data (list) : list of cars data

    Return
    ads_data (list)
    """
    ads_data = []
    for i in range(1, n+1):
        ad_id = i
        seller_id = random.choice(user_rows)['user_id']
        product_id = random.choice(cars['product_id'])
        color = random.choice([ "Silver", "Putih","Merah", "Hitam", "Biru", "Abu-abu", "Hijau", "Oranye", "Kuning"])
        mileage = random.randint(10000, 80000)
        price= random.randint(800, 4000)*100000
        negotiable = random.choice([True, False])
        is_sold = random.choice([True, False])
        desc = random.choice(['No minus','lecet sedikit','tanpa lecet','pajak hidup',""])
        post_date = fake.date_time_between_dates(datetime_start =datetime.strptime(user_rows[seller_id-1]['date_created'], '%Y-%m-%d %H:%M:%S'), datetime_end =datetime(2023, 12, 31, 23, 59, 59)).strftime("%Y-%m-%d %H:%M:%S")
        ads_data.append({
            "ad_id" : ad_id,
            "seller_id" : seller_id,
            "product_id" : product_id,
            "color": color,
            "mileage" : mileage,
            "ads_price":price,
            "description" : desc,
            "negotiable" : negotiable,
            "date_posted_ads" : post_date,
            "is_sold":is_sold
        })

    return ads_data

#define columns for ads table
ad_cols = ["ad_id", "seller_id", "product_id", "color", "mileage", "ads_price","description", "negotiable", "date_posted_ads", "is_sold"]
#generate rows for ads table
ad_rows = generate_dummy_ads(n_ads,user_rows)
ads_df=pd.DataFrame(ad_rows, columns=ad_cols)
ads_df
```

```python
def generate_dummy_bids(n,user_rows,ad_rows):
    """
    A function for generating bids data.

    Args
    users_rows (list) : list of user rows that has generated previously
    ad_rows (list) : list of ad rows that has generated previously

    Return
    bids_data (list)
    """
    bids_data = []
    for i in range(1, n+1):
        bid_id = i
        ads_id = random.choice(ad_rows)['ad_id']
        buyer_id = random.choice(user_rows)['user_id']
        #Buyer_id tidak boleh sama dengan seller_id
        while buyer_id==ad_rows[ads_id-1]['seller_id']:
            buyer_id = random.choice(user_rows)['user_id']
        
        #Jika negotiable di ads=False maka harganya sama dengan harga ads
        if ad_rows[ads_id-1]['negotiable']==True:
            price= random.randint(80000000, ad_rows[ads_id-1]['ads_price'])
        else:
            price=ad_rows[ads_id-1]['ads_price']
        #Tanggal bid harus lebih besar dari tanggal ads
        bid_date = fake.date_time_between_dates(datetime_start =datetime.strptime(ad_rows[ads_id-1]['date_posted_ads'], '%Y-%m-%d %H:%M:%S'), datetime_end =datetime(2023, 12, 31, 23, 59, 59)).strftime("%Y-%m-%d %H:%M:%S")
        bid_status= random.choice(['Sent'])
        bids_data.append({
            "bid_id" : bid_id,
            "ad_id" : ads_id,
            "buyer_id": buyer_id,
            "bid_price":price,
            "bid_date" : bid_date,
            "bid_status":bid_status
        })

    return bids_data

#define columns for ads table
bid_cols = ["bid_id", "ad_id", "buyer_id", "bid_price", "bid_date","bid_status"]
#generate rows for ads table
bid_rows = generate_dummy_bids(5000,user_rows,ad_rows)
bid_rows
```

```python
#Merubah bid_status pada ads yang sudah terjual

sold_ads=[]
for i in range(len(ad_rows)):
    if ad_rows[i]['is_sold']==True:
               sold_ads.append(ad_rows[i]['ad_id'])
win_bid=[]
already_have_winner=[]
for x in bid_rows:
    if x['ad_id'] in sold_ads and x['ad_id'] not in already_have_winner:
      x['bid_status']='Succeed'
      already_have_winner.append(x['ad_id'])
      win_bid.append(x)
    elif x['ad_id'] in sold_ads :
      x['bid_status']='Rejected'
      
  

```

```python

def generate_dummy_deals(win_bid):
    """
    A function for generating deals data.

    Args
    win_bid (list) : list of user rows that has generated previously

    Return
    deals_data (list)
    """
    deals_data = []
    i=0
    for x in win_bid:
        deal_id = i+1
        ads_id=x['ad_id']
        bid_id=x['bid_id']
        payment=i+1
        i+=1
        deals_data.append({
            "deal_id" : deal_id,
            "ad_id" : ads_id,
            "bid_id" : bid_id,
            "payment_id" : payment
        })


    return deals_data

#define columns for deals table
deal_cols = ["deal_id","ad_id", "bid_id", "payment_id"]
#generate rows for deals table
deal_rows = generate_dummy_deals(win_bid)
deal_df=pd.DataFrame(deal_rows, columns=deal_cols)
deal_df
```

```python


def generate_dummy_payment(win_bid):
    """
    A function for generating payment data.

    Args
    win_bid (list) : list of user rows that has generated previously

    Return
    payment_data (list)
    """
    payment_data=[]
    i=0
    for i in range(len(win_bid)):
        payment=i+1
        payment_type = random.choice([ "VISA","Credit Card","Bank trasfer","Virtual Account"])
        payment_date = fake.date_time_between_dates(datetime_start =datetime.strptime(win_bid[i]['bid_date'], '%Y-%m-%d %H:%M:%S'), datetime_end =datetime(2023, 12, 31, 23, 59, 59)).strftime("%Y-%m-%d %H:%M:%S")
        i+=1


        payment_data.append({
            "payment_id" : payment,
            "payment_type":payment_type,
            "payment_date":payment_date
        })
    return payment_data

#define columns for payment table
payment_cols = ["payment_id","payment_type", "payment_date"]
#generate rows for payment table
payment_rows = generate_dummy_payment(win_bid)
payment_df=pd.DataFrame(payment_rows, columns=payment_cols)
payment_df
```
